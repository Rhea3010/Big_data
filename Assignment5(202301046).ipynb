```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_data, x='month_number', y='total_profit', marker='o')
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

```

```python

plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_data.drop(columns=['month_number', 'total_profit']))
plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend(title='Product', loc='upper left')
plt.grid(True)
plt.show()

```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=sales_data.drop(columns=['month_number']))
plt.title('Bar Chart of Sales Data Features')
plt.ylabel('Value')
plt.xlabel('Features')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)


scatter_plot = px.scatter(sales_data, x='total_units', y='total_profit',
                          color='month_number', size='total_profit',
                          title='Scatter Plot of Total Units vs Total Profit',
                          labels={'total_units': 'Total Units', 'total_profit': 'Total Profit', 'month_number': 'Month Number'})

scatter_plot.show()
```

```python
import pandas as pd
import plotly.express as px


url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)


monthly_profit = sales_data.groupby('month_number')['total_profit'].sum().reset_index()


bar_chart = px.bar(monthly_profit, x='month_number', y='total_profit',
                   title='Monthly Total Profit',
                   labels={'month_number': 'Month Number', 'total_profit': 'Total Profit'})


bar_chart.show()
```

```python
import pandas as pd
import plotly.express as px


url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
sales_data = pd.read_csv(url)


histogram = px.histogram(sales_data, x='total_profit', nbins=30,
                         title='Histogram of Total Profit',
                         labels={'total_profit': 'Total Profit'})


histogram.show()
```
